```python
import numpy as np
import matplotlib.pyplot as plt

resolution = 100
gArray = np.linspace(-1, 1, resolution)
allE = []           

def hamiltonian(g):
    H = np.array([
        [2-g, -g/2, -g/2, -g/2, -g/2, 0],
        [-g/2, 4-g, -g/2, -g/2, 0, -g/2],
        [-g/2, -g/2, 6-g, 0, -g/2, -g/2],
        [-g/2, -g/2, 0, 6-g, -g/2, -g/2],
        [-g/2, 0, -g/2, -g/2, 8-g, -g/2],
        [0, -g/2, -g/2, -g/2, -g/2, 10-g]
    ])
    return np.linalg.eigh(H)

for g in gArray:
    E, psi = hamiltonian(g)
    allE.append(E)

allE = np.array(allE)   

plt.figure()

for n in range(allE.shape[1]):
    plt.plot(gArray, allE[:, n], label=fr"$\Phi_{n}$")

plt.xlabel("g")
plt.ylabel("Energy eigenvalues")
plt.title("Spectrum of H as function of g")
plt.legend(fontsize=10)
plt.show()
```

```python
plt.plot(gArray, allE[:, 0])
plt.xlabel("g")
plt.ylabel("Ground state energy")
```

```python
resolution = 100
gArray = np.linspace(-1, 1, resolution)
allE_trun = []           

def hamiltonian_t(g):
    H = np.array([
        [2-g, -g/2, -g/2, -g/2, -g/2],
        [-g/2, 4-g, -g/2, -g/2, 0],
        [-g/2, -g/2, 6-g, 0, -g/2],
        [-g/2, -g/2, 0, 6-g, -g/2],
        [-g/2, 0, -g/2, -g/2, 8-g],
    ])
    return np.linalg.eigh(H)

for g in gArray:
    E, psi = hamiltonian_t(g)
    allE_trun.append(E)

allE_trun = np.array(allE_trun)   

plt.figure()

for n in range(allE_trun.shape[1]):
    plt.plot(gArray, allE_trun[:, n], label=fr"$\Phi_{n}$")

plt.xlabel("g")
plt.ylabel("Energy eigenvalues")
plt.title("Spectrum of H as function of g")
plt.legend()
plt.show()
```

Comparison between H and truncated H in terms of ground state $|\Phi_0 \rangle$

```python
import numpy as np
import matplotlib.pyplot as plt

gArray = np.linspace(-1, 1, 100)

allE_full = []
allE_trunc = []

for g in gArray:
    E_full, _  = hamiltonian(g)     
    E_trunc, _ = hamiltonian_t(g)    
    allE_full.append(E_full)
    allE_trunc.append(E_trunc)

allE_full  = np.array(allE_full)   
allE_trunc = np.array(allE_trunc)  


E0_full  = allE_full[:, 0]         
E0_trunc = allE_trunc[:, 0]

# deviation of truncated from exact
deltaE = E0_trunc - E0_full

plt.figure()
plt.axhline(0.0, color="k", linewidth=1, linestyle="--", label="Exact H baseline")
plt.plot(gArray, deltaE, label=r"$E_0^{\mathrm{trunc}} - E_0^{\mathrm{exact}}$")

plt.xlabel("g")
plt.ylabel(r"Energy deviation $\Delta E$")
plt.title("Deviation of truncated ground state from exact ground state", fontsize=14)
plt.legend(fontsize=14)
plt.show()
```
