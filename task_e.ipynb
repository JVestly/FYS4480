Diagonalisation code for (e)

```python
import numpy as np

# For the first parameter set
V = -1/3
W = -1/4
H = np.array([
    [-2*2, 0, np.sqrt(6)*V, 0, 0],
    [0, -2+3*W, 0, 3*V, 0],
    [np.sqrt(6)*V, 0, 4*W, 0, np.sqrt(6)*V],
    [0, 3*V, 0, 2+3*W, 0],
    [0, 0, np.sqrt(6)*V, 0, 2*2]
])

E, psi = np.linalg.eigh(H)  # eigenvalues and eigenfunctions
print(E)
print(psi)
```

```python
# For the second parameter set
V = -4/3
W = -1
H = np.array([
    [-2*2, 0, np.sqrt(6)*V, 0, 0],
    [0, -2+3*W, 0, 3*V, 0],
    [np.sqrt(6)*V, 0, 4*W, 0, np.sqrt(6)*V],
    [0, 3*V, 0, 2+3*W, 0],
    [0, 0, np.sqrt(6)*V, 0, 2*2]
])

E, psi = np.linalg.eigh(H)  # eigenvalues and eigenfunctions
print(E)
print(psi)
```
